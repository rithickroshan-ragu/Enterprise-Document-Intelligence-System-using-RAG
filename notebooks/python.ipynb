```python
from pypdf import PdfReader

reader = PdfReader("The indigenous badugar of the Nilgiris.pdf")
```

```python
print(reader.pages)
text = ""
for page in reader.pages:
    text += page.extract_text() + "\n"
print(text)
```

```python
N = len(text)
print(N)
chunk_size = 100
for i in range(0, N, chunk_size):
    chunk = text[i:i + chunk_size]
    print(chunk)

```
